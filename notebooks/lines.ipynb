```python
import math
from itertools import islice, batched, product

from matplotlib import pyplot as plt
import numpy as np
import shapely
from shapely import plotting
import random
```

```python
x_0 = 0
y_0 = 0
# Actually, points. Lines are created by zipping the points together
lines = []
for i in range(0, 40):
    d = random.randint(10,135)
    # print(d)
    theta = math.radians(d)
    x = math.cos(theta) + x_0
    y = math.sin(theta) + y_0
    lines.append((x,y))
    h = random.uniform(0.6, 1.0)
    # print(x, y, h)
    x_0 = x
    y_0 = y

# s = shapely.LineString(lines)

fig, ax = plt.subplots()
combo = zip([(0,0)] + lines[:-1], lines)
# print(combo)
for l in combo:
    # print(l)
    s = shapely.LineString(l)
    # print(s)
    plotting.plot_line(shapely.LineString(l), ax=ax)
```
