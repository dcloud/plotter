## Hypotrochoid

<https://mathworld.wolfram.com/Hypotrochoid.html>


$$x = (a - b)\cos\left(\theta\right) - b\cos\left( \frac{a+b}{b}\theta \right)$$

$$y = (a - b)\sin\left(\theta\right) - b\sin\left( \frac{a+b}{b}\theta \right)$$


### Numpy & Matplotlib

```python
radius_a = 40
radius_b = 10
h = radius_a - radius_b/10
```

```python
theta = np.arange(360)
```

```python
x = (radius_a - radius_b) * np.cos(theta) + h*np.cos((radius_a - radius_b)/radius_b * theta)
y = (radius_a - radius_b) * np.sin(theta) - h*np.sin((radius_a - radius_b)/radius_b * theta)
```

```python
fig, ax = plt.subplots()
ax.set_box_aspect(1)
ax.plot(x,y, '.r')
plt.show()
```

### Shapely

```python
import math
```

```python
def hypotrochoid(degrees=range(361), a=40, b=10, h=lambda a=2,b=1: a-b):
    thetas = (math.radians(x) for x in degrees)
    pts = []
    h_val = h(a,b)
    for t in thetas:
        x = (a - b) * math.cos(t) + h_val*math.cos((a - b)/b * t)
        y = (a - b) * math.sin(t) - h_val*math.sin((a - b)/b * t)
        pts.append([x,y])
    return pts
```

```python
shape_list = hypotrochoid(a=60, b=10, h=lambda a,b: (a-b)*2)
s = shapely.LineString(shape_list)
display(s)

shape_list = hypotrochoid(a=50, b=10, h=lambda a,b: (a-b)*2)
s = shapely.LineString(shape_list)
display(s)
```

```python
shape_list = hypotrochoid(a=60, b=10, h=lambda a,b: (a-b)*2, degrees=range(0,361, 8))
s = shapely.LineString(shape_list)
display(s)

shape_list = hypotrochoid(a=50, b=10, h=lambda a,b: (a-b)*2, degrees=range(0,361, 9))
s = shapely.LineString(shape_list)
display(s)

shape_list = hypotrochoid(a=135, b=15, h=lambda a,b: (a-b)*2, degrees=range(0,361, 8))
s = shapely.LineString(shape_list)
display(s)

shape_list = hypotrochoid(a=135, b=15, h=lambda a,b: (a-b)*2, degrees=range(0,361, 10))
s = shapely.LineString(shape_list)
display(s)

shape_list = hypotrochoid(a=105, b=15, h=lambda a,b: (a-b)*2, degrees=range(0,361, 5))
s = shapely.LineString(shape_list)
display(s)
```

```python
shape_list = hypotrochoid(a=90, b=15, h=lambda a,b: b)
s = shapely.LineString(shape_list)
display(s)
```

## Epicycloid

<https://mathworld.wolfram.com/Epicycloid.html>

$$x = (a + b)\cos\left(\theta\right) - b\cos\left( \frac{a+b}{b}\theta \right)$$

$$y = (a + b)\sin\left(\theta\right) - b\sin\left( \frac{a+b}{b}\theta \right)$$


```python
def epicycloid(degrees=range(361), a=40, b=10, h=None):
    thetas = (math.radians(x) for x in degrees)
    if not h:
        h = b
    pts = []
    for t in thetas:
        x = (a + b) * math.cos(t) - b*math.cos((a+b)/b * t)
        y = (a + b) * math.sin(t) - b*math.sin((a+b)/b * t)
        pts.append([x, y])
    return pts
```

```python
shape_list = epicycloid(a=120, b=10, h=40)
s = shapely.LineString(shape_list)
display(s)
```

```python
shape_list = epicycloid(a=120, b=5, degrees=range(91))
s = shapely.LineString(shape_list)
display(s)

shape_list = epicycloid(a=120, b=6, degrees=range(361))
s = shapely.LineString(shape_list)
display(s)

shape_list = epicycloid(a=120, b=6, degrees=range(0, 361, 5))
s = shapely.LineString(shape_list)
display(s)

shape_list = epicycloid(a=120, b=6, degrees=range(0, 361, 8))
s = shapely.LineString(shape_list)
display(s)


shape_list = epicycloid(a=180, b=30, degrees=range(0, 361, 10))
s = shapely.LineString(shape_list)
display(s)


shape_list = epicycloid(a=180, b=30, degrees=range(0, 361, 15))
s = shapely.LineString(shape_list)
display(s)
```

## Spirals

```python
def spiral(degrees=range(361), a=0, b=math.pi*2):
    thetas = (math.radians(x) for x in degrees)
    pts = []
    for t in thetas:
        x = (a + b*t) * math.cos(t)
        y = (a + b*t) * math.sin(t)
        pts.append([x, y])
    return pts
```

```python
s = shapely.LineString(spiral())
display(s)

s = shapely.LineString(spiral(degrees=range(360*4 + 1)))
display(s)

s = shapely.LineString(spiral(degrees=range(360*6 + 1), a=90))
display(s)

s = shapely.LineString(spiral(degrees=range(360*3+1), a=45))
display(s)
```

#### Limiting the steps

```python
s = shapely.LineString(spiral(degrees=range(0, 360*6+1, 60)))
display(s)

s = shapely.LineString(spiral(degrees=range(0, 360*10+1, 75)))
display(s)

s = shapely.LineString(spiral(degrees=range(0, 360*20+1, 75)))
display(s)

s = shapely.LineString(spiral(degrees=range(0, 360*6+1, 90), a=90))
display(s)


s = shapely.LineString(spiral(degrees=range(0, 360*6+1, 45), a=45, b=math.pi))
display(s)
```
